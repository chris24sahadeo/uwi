## Python Implementation example #1 from Lab using [`sklearn.cluster.DBSCAN`](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html)

### Imports

```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

### Random data generation with [`sklearn.datasets.make_blobs`](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html)

```python
# randomly creating data points
X, y = make_blobs(n_samples=200, centers=5)
print(X.shape, y.shape)     # give us (number of rows, number of features)
```

### Making and training the model

```python
# eps is neighbourhood distance
# min_samples is the min elements per neighborhood
clf = DBSCAN(eps=1, min_samples=5).fit(X)
print(clf.labels_)
```

### Plotting with [`matplotlib.pyplot`](https://matplotlib.org/tutorials/index.html)

```python
# plotting
plt.figure(figsize=(15, 10))
plt.grid(True)
plt.scatter(X[:, 0], X[:, 1], c = clf.labels_)
plt.show()
```

## Python Implementation example #2

### Defining some constants

```python
FIG_HEIGHT = 15
FIG_WIDTH = 15
MARKER_SIZE = 200
FILE_NAME = 'testSet.csv'
```

### Imports


```python
import numpy as np                      # for arrays
import matplotlib.pyplot as plt         # for plotting
from sklearn.cluster import DBSCAN      # for running the KMeans algorithm
```

### Reading Data from `testSet.csv`
This data was made using [`sklearn.datasets.make_blobs`](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html) with `n_samples`=20 and `centers`=4

```python
print('Reading data from {}...'.format(FILE_NAME))
try:
    X = np.genfromtxt(FILE_NAME, delimiter=',')
except IOError:
    print('Failed to read file...')

print('Printing Data set: ')
print(X)
```

### Visualizing data before DBSCAN

```python
plt.figure(figsize=(FIG_WIDTH, FIG_HEIGHT))
plt.grid(True)
plt.scatter(X[:,0], X[:,1])
plt.title('Data Set')
plt.show()
```

### Performing DBSCAN

```python
clf = DBSCAN(eps=2, min_samples=3).fit(X)
print('Cluster Assignments:\n{}\n'.format(clf.labels_))
```

### Visualizing data after DBSCAN

```python
plt.figure(figsize=(FIG_WIDTH, FIG_HEIGHT))
plt.grid(True)
plt.scatter(X[:,0], X[:,1], c=clf.labels_)
plt.title('Data Set after DBSCAN')
plt.show()
```
